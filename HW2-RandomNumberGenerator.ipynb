# Random number generator from a range, using PI

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### File containing million digits of the number pi ("PI_digits.txt")


```python
with open("PI_digits.txt", 'r') as f:
    PI = f.read()
```

```python
def random_number(lowerbound, upperbound):
    global position
    length = int(PI[position])
    if length == 0:
        length = 1
    number = int(PI[(1+position):(1+position+length)])
    position += (length+1)
    
    return number % (upperbound - lowerbound + 1) + lowerbound 
```

```python
how_many = 1000
position = 9 # initially position is the SEED
lowerbound = 4
upperbound = 50

arr = np.array([])
for i in range(how_many):
    arr = np.append(arr, random_number(lowerbound, upperbound))
```

```python
print(arr)
```

```python
x = np.arange(0, how_many, 1)
y = arr[x]

fig, ax = plt.subplots()
ax.scatter(x, y)

ax.set(xlabel='Index of generated number', ylabel='The random number',
       title='Random numbers generated from PI')
ax.grid()
plt.show()
```
